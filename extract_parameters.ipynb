```python
with open('content/vehicle_depth.json', 'r') as file:
    vehicle_depth = json.load(file)
    #print(vehicle_depth)
    #1. frame num 2. x, y, z
print(vehicle_depth['vehicle0'])
```

```python
how_far1 = []
how_far2 = []
for frame in vehicle_depth['vehicle0']:
    how_far1.append(frame[1][2])
    how_far2.append(frame[2][2])
print(how_far1)
print(how_far2)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.plot(how_far1, marker='o', linestyle='-', color='b')  # 선 그래프
plt.plot(how_far2, marker='o', linestyle='-', color='r')  # 선 그래프
plt.title('Data Visualization')  # 그래프 제목
plt.xlabel('Index')  # x축 라벨
plt.ylabel('Value')  # y축 라벨
plt.grid(True)  # 그리드 추가
plt.show()  # 그래프 출력
```
